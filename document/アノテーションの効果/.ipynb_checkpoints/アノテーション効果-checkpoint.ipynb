# アノテーション効果検証レポート

1枚の画像に対して、複数認識対象がある画像に対して、ある１対象のみをアノテーションする場合と、全対象をアノテーションした場合での学習結果、及びテスト結果を表す。

```python
import matplotlib.pyplot as plt
from PIL import Image
```

## 学習結果

```python
matrix_100_0623 = Image.open('資料/confusionMatrix_100_20200623.png')
matrix_100_0723 = Image.open('資料/confusionMatrix_100_20200723.png')
matrix_1000_0723 = Image.open('資料/confusionMatrix_1000_20200723.png')

plt.figure(figsize=(100,100))

plt.subplot(2,2,1)
plt.imshow(matrix_100_0623)
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(matrix_100_0723)
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(matrix_1000_0723)
plt.axis('off')
```
