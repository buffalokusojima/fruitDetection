```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Yu Gothic']
from PIL import Image

from src.yolo import YOLO

import os
```

```python
def detect_img(yolo):
     
    try:
        img = input('Input image filename:')
        img = os.path.join('画像', img)
        image = Image.open(img)
    except:
        print('Open Image Error:', img)
        return None, None
    else:
        r_image, result = yolo.detect_image(image)
        
    return r_image, result
```

```python
model = YOLO(model_path='モデル/0.99-model.h5', classes_path='モデル/class.txt', anchors_path='モデル/yolo_anchors.txt')
```

```python
image, result = detect_img(model)
plt.imshow(image)
```
