# テストデータをランダムにフォルダから抽出して作成する

## モジュールインポート

```python
import os, glob, shutil, random
```

## 定数宣言

```python
target_dir = '../train_set/学習データ/479削除データ'
output_dir = '../train_set/テストデータ'
class_file = 'type_class.txt'
fileNum = 100
```

```python
img_path = os.path.join(target_dir, 'image')
test_dir = os.path.join(output_dir, 'random_'+str(fileNum))
if not os.path.isdir(test_dir):
    print("foder made:",test_dir)
    os.mkdir(test_dir)
```

## クラスファイル取得

```python
class_list = []
with open(class_file, 'r') as f:
    line = f.readline()
    while line:
        class_list.append(line.replace("\n", ""))
        line = f.readline()
```

```python
print(class_list)
```

## データをコピー

```python
img_list =  glob.glob(os.path.join(target_dir, '**', '*.jpg'), recursive=True)
random.shuffle(img_list)
img_list = img_list[:100]
```

```python
if not os.path.isdir(test_dir):
    print("folder made:", test_dir)
    os.mkdir(test_dir)
    os.mkdir(os.path.join(test_dir, "image"))
    os.mkdir(os.path.join(test_dir, "annotation"))
    
test_list = []
for img in img_list:
    img_name = img.split("/")[-1]
    output_img = os.path.join(test_dir, "image", img_name)
    shutil.copyfile(img, output_img)
    print("copy", img, "to", output_img)
    
    annotation = img.replace("image", "annotation").replace(".jpg", ".txt")
    annotation_name = annotation.split("/")[-1]
    output_annotation = os.path.join(test_dir, "annotation", annotation_name)
    shutil.copyfile(annotation, output_annotation)
    print("copy", annotation, "to", output_annotation)
    
    with open(output_annotation, 'r') as f:
        line = f.readline()
        while line:
            annotation = line
            line = f.readline()
    
    class_index = None
    for c in class_list:
        if img.find(c) > 0:
            class_index = str(class_list.index(c))
            
    test_list.append([output_img, annotation, class_index])
    
last = test_list[-1]
test_file = os.path.join(test_dir, 'test.txt')
with open(test_file, 'w', encoding='utf-8') as f:
    for test in test_list:
        jpg = test[0]
        f.write(jpg+ ' ')
        test[1] = test[1].split(" ")
        test[1] = ",".join(test[1])
        f.write(test[1])
        f.write(","+test[2])
        if test != last:
            f.write("\n")
print("file made:", test_file)
```
