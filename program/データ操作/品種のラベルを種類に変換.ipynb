```python
import os,glob,shutil
from xml.etree import ElementTree
```

```python
target_dir = "../学習データ/混合データ/品種"
```

```python
target_path = os.path.join(target_dir, "annotation")
```

```python
count = 0
error_list = []
for file in glob.glob(os.path.join(target_path, "**", "*.xml"), recursive=True):
    output_file = file.replace("品種", "種類")
    output_dir = "/".join(output_file.split("/")[:-1])
    
    
    type_class = file.split("/")[-3].split("_")[1]
    
    if not os.path.isdir(output_dir):
        print("folder made:", output_dir)
        os.makedirs(output_dir)
        
    #shutil.copyfile(file, output_file)
        
    try:
        tree = ElementTree.parse(file)
        
        elem = tree.getroot()
        
        objects = elem.findall("object")
        
        for obj in objects:
            obj.find("name").text = type_class
            
        tree.write(output_file, encoding="UTF-8")
    except Exception as e:
        error_list.append(e)
    print(output_file)
    count += 1
    
print("file making done:", count)
    
print("error:",len(error_list))
print(error_list)
```
