# labelimgで作成したアノテーションをkeras-yolo3用に変換する

## モジュールインポート

```python
import os,glob
from xml.etree import ElementTree
```

## 定数宣言

```python
folder_type = "品種"

"""
種類

target_dir = '../学習データ/追加データ(keras-yolo3)/種類/annotation'
output_dir = '../学習データ/追加データ(keras-yolo3)/種類/annotation'
classes_path = '../本番/class/種類クラス/class.txt'


品種
"""
target_dir = '../学習データ/混合データ/品種/annotation/2_梨/'
output_dir = '../学習データ/混合データ/品種/annotation/'

classes_path = '../本番/class/品種クラス/2_梨/class.txt'
```

## クラスファイル取得

```python
class_list = []
with open(classes_path, 'r', encoding='UTF-8') as f:
    line = f.readline()
    while line:
        class_list.append(line.replace("\n",""))
        line = f.readline()
print(class_list)
```

## フォルダを回してxmlファイルから必要な情報を抽出したtxtファイルを作成する

```python
file_count = 0
error_list = []

for folder in glob.glob(os.path.join(target_dir,'*')):
    
    if not os.path.isdir(folder):
        continue
    
    if folder_type is "種類":
        folder_number = -3
    else:
        folder_number = -2
    
    for path in glob.glob(os.path.join(folder, "**", "*.xml"), recursive=True):
        
        try:
            tree = ElementTree.parse(path)
            elem = tree.getroot()
            folder = elem.findtext("path")
            type_folder = folder.split("\\")[-3]
            veriety_folder = folder.split("\\")[-2]

            output_folder = os.path.join(output_dir, type_folder, veriety_folder)

            if not os.path.isdir(output_folder):
                os.makedirs(output_folder)
                print("folder made:", output_folder)

            objects = elem.findall("object")

            annotation_list = []

            for obj in objects:


                label = class_list.index(obj.find("name").text)
                for bbox in obj.findall("bndbox"):
                    xmin = bbox.find("xmin").text
                    xmax = bbox.find("xmax").text
                    ymin = bbox.find("ymin").text
                    ymax = bbox.find("ymax").text
                    annotation_list.append([xmin, ymin, xmax, ymax, str(label)])



            annotation_file = os.path.join(output_folder, path.split("\\")[-1].replace("xml","txt"))
            
            with open(annotation_file, 'w') as f:
                last = annotation_file[-1]
                for annotation in annotation_list:
                    f.write(",".join(annotation))
                    if not annotation is last:
                        f.write("\n")
                print("writing:", annotation_file)
            file_count += 1
        except:
            error_list.append(path)

print("file making done:", file_count)
print("error list:", error_list)
```
