```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from PIL import Image, ImageOps, ImageEnhance
```

```python
inputImagePath = r'\\172.18.21.187\aiコンテスト2020\100_開発\テストデータ\とりあえず\image\1_リンゴ\1_あいかの香り\0.jpg'
print(os.path.exists(inputImagePath))
```

```python
# original
```

```python
img = Image.open(inputImagePath)
plt.imshow(img)
```

```python
# 彩度を上げる
```

```python
sat_up = ImageEnhance.Color(img)
sat_up_image = sat_up.enhance(1.5)
plt.imshow(sat_up_image)
```

```python
# 彩度を下げる
```

```python
sat_down = ImageEnhance.Color(img)
sat_down_image = sat_down.enhance(0.5)
plt.imshow(sat_down_image)
```

```python
# コントラストを上げる
```

```python
cont_up = ImageEnhance.Contrast(img)
cont_up_image = cont_up.enhance(1.5)
plt.imshow(cont_up_image)
```

```python
# コントラストを下げる
```

```python
cont_down = ImageEnhance.Contrast(img)
cont_down_image = cont_down.enhance(0.5)
plt.imshow(cont_down_image)
```

```python
# 明度を上げる
```

```python
bri_up = ImageEnhance.Brightness(img)
bri_up_image = bri_up.enhance(1.5)
plt.imshow(bri_up_image)
```

```python
# 明度を下げる
```

```python
bri_down = ImageEnhance.Brightness(img)
bri_down_image = bri_down.enhance(0.5)
plt.imshow(bri_down_image)
```

```python
# シャープネスを上げる
```

```python
shr_up = ImageEnhance.Sharpness(img)
shr_up_image = shr_up.enhance(10)
plt.imshow(shr_up_image)
```

```python
# シャープネスを下げる
```

```python
shr_down = ImageEnhance.Sharpness(img)
shr_down_image = shr_down.enhance(0.1)
plt.imshow(shr_down_image)
```

```python
# 拡大縮小
```

```python
ratio = 2.0
w, h = img.size
img_resize = img.resize((int(w * ratio), int(h * ratio)))
```

```python
plt.imshow(img_resize)
```

```python
# 反転
```

```python
# 上下反転
```

```python
flip_image = ImageOps.flip(img)
plt.imshow(flip_image)
```

```python
# 左右反転
```

```python
mirror_image = ImageOps.mirror(img)
plt.imshow(mirror_image)
```

```python
# 回転
```

```python
rotate_90_image = img.rotate(90, expand = True)
plt.imshow(rotate_90_image)
```

```python
rotate_270_image = img.rotate(270, expand = True)
plt.imshow(rotate_270_image)
```
