```python
def changeSaturation(originalImage, ratio, name, variety, fruit):
    saturation = ImageEnhance.Color(originalImage)
    changeSaturationImage = saturation.enhance(ratio)
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")
    
    
    outputPath = os.path.join(outputRoot, "彩度" + str(ratio), "種類", "image", fruit, variety, name)
    outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    changeSaturationImage.save(outputPath)
    
    # txtからデータをとってきてる
    coorList = []
    if os.path.exists(inputTextPath):
        with open(inputTextPath) as f:
            for s_line in f:
                if len(s_line.split(',')) == 1:
                    return
                coorList.append(s_line.split(','))
    
    if os.path.exists(inputTextPath):
        shutil.copy(inputTextPath, outputTextPath)
        make_xml(coorList, outputTextPath)
    
    
    
    

def changeContrast(originalImage, ratio, name, variety, fruit):
    contrast = ImageEnhance.Contrast(originalImage)
    changeContrastImage = contrast.enhance(ratio)
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")

    outputPath = os.path.join(outputRoot, "コントラスト" + str(ratio), "種類", "image", fruit, variety, name)
    outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    if os.path.exists(inputTextPath):
        shutil.copy(inputTextPath, outputTextPath)
    changeContrastImage.save(outputPath)
    
    

def changeBrightness(originalImage, ratio, name, variety, fruit):
    brightness = ImageEnhance.Brightness(originalImage)
    changeBrightnessImage = brightness.enhance(ratio)
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")
    
    outputPath = os.path.join(outputRoot, "明度" + str(ratio), "種類", "image", fruit, variety, name)
    outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    if os.path.exists(inputTextPath):
        shutil.copy(inputTextPath, outputTextPath)
    changeBrightnessImage.save(outputPath)
    
    

def changeSharpness(originalImage, ratio, name, variety, fruit):
    sharpness = ImageEnhance.Sharpness(originalImage)
    changeSharpnessImage = sharpness.enhance(ratio)
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")
    
    outputPath = os.path.join(outputRoot, "シャープネス" + str(ratio), "種類", "image", fruit, variety, name)
    outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    if os.path.exists(inputTextPath):
        shutil.copy(inputTextPath, outputTextPath)
    changeSharpnessImage.save(outputPath)
    
    

def changeSize(originalImage, ratio, name, variety, fruit):
    w, h = originalImage.size
    resizedImage = originalImage.resize((int(w * ratio), int(h * ratio)))
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")
    
    coorList = []
    if os.path.exists(inputTextPath):
        with open(inputTextPath) as f:
            for s_line in f:
                if len(s_line.split(',')) == 1:
                    return
                coorList.append(s_line.split(','))
    newCoorList = []
    for oneCoorList in coorList:
        xmin = int(oneCoorList[0])
        ymin = int(oneCoorList[1])
        xmax = int(oneCoorList[2])
        ymax = int(oneCoorList[3])
        newCoorList.append([str(int(xmin * ratio)),str(int(ymin * ratio)), str(int(xmax * ratio)), str(int(ymax * ratio)), oneCoorList[4]])
    
    outputPath = os.path.join(outputRoot, "大きさ" + str(ratio), "種類", "image", fruit, variety, name)
    outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    with open(outputTextPath, mode = 'w') as f:
        for oneCoorList in newCoorList:
            f.write(oneCoorList[0] + "," + oneCoorList[1] + "," + oneCoorList[2] + "," + oneCoorList[3] + "," + oneCoorList[4])
    resizedImage.save(outputPath)
    
    

def flipVertical(originalImage, name, variety, fruit):
    flipedImage = ImageOps.flip(originalImage)
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")
    
    coorList = []
    if os.path.exists(inputTextPath):
        with open(inputTextPath) as f:
            for s_line in f:
                if len(s_line.split(',')) == 1:
                    return
                coorList.append(s_line.split(','))
    newCoorList = []
    w, h = originalImage.size
    for oneCoorList in coorList:
        xmin = int(oneCoorList[0])
        ymin = int(oneCoorList[1])
        xmax = int(oneCoorList[2])
        ymax = int(oneCoorList[3])
        newCoorList.append([str(xmin), str(h - ymax), str(xmax), str(h - ymin), oneCoorList[4]])
    
    
    outputPath = os.path.join(outputRoot, "上下反転", "種類", "image", fruit, variety, name)
    outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    with open(outputTextPath, mode = 'w') as f:
        for oneCoorList in newCoorList:
            f.write(oneCoorList[0] + "," + oneCoorList[1] + "," + oneCoorList[2] + "," + oneCoorList[3] + "," + oneCoorList[4])
    flipedImage.save(outputPath)
    
    

def flipHorizontal(originalImage, name, variety, fruit):
    mirrorImage = ImageOps.mirror(originalImage)
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")
    
    coorList = []
    if os.path.exists(inputTextPath):
        with open(inputTextPath) as f:
            for s_line in f:
                if len(s_line.split(',')) == 1:
                    return
                coorList.append(s_line.split(','))
                
    newCoorList = []
    w, h = originalImage.size
    for oneCoorList in coorList:
        xmin = int(oneCoorList[0])
        ymin = int(oneCoorList[1])
        xmax = int(oneCoorList[2])
        ymax = int(oneCoorList[3])
        newCoorList.append([str(w - xmax), str(ymin), str(w - xmin), str(ymax), oneCoorList[4]])
    
    outputPath = os.path.join(outputRoot, "左右反転", "種類", "image", fruit, variety, name)
    outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    with open(outputTextPath, mode = 'w') as f:
        for oneCoorList in newCoorList:
            f.write(oneCoorList[0] + "," + oneCoorList[1] + "," + oneCoorList[2] + "," + oneCoorList[3] + "," + oneCoorList[4])
    mirrorImage.save(outputPath)
    
    

def rotation(originalImage, degrees, name, variety, fruit):
    rotatedImage = originalImage.rotate(degrees, expand = True)
    inputTextPath = os.path.join(inputTextFolderPath, variety, name).replace("image", "annotation").replace(".jpg", ".txt")
    
    coorList = []
    if os.path.exists(inputTextPath):
        with open(inputTextPath) as f:
            for s_line in f:
                if len(s_line.split(',')) == 1:
                    return
                coorList.append(s_line.split(','))
    newCoorList = []
    w, h = originalImage.size
    
    if degrees == 90:
        for oneCoorList in coorList:
            xmin = int(oneCoorList[0])
            ymin = int(oneCoorList[1])
            xmax = int(oneCoorList[2])
            ymax = int(oneCoorList[3])
            newCoorList.append([str(ymin), str(w - xmax), str(ymax), str(w - xmin), oneCoorList[4]])
        outputPath = os.path.join(outputRoot, "回転" + str(degrees), "種類", "image", fruit, variety, name)
        outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
        
    elif degrees == 270:
        for oneCoorList in coorList:
            xmin = int(oneCoorList[0])
            ymin = int(oneCoorList[1])
            xmax = int(oneCoorList[2])
            ymax = int(oneCoorList[3])
            newCoorList.append([str(h - ymax), str(xmin), str(h - ymin), str(xmax), oneCoorList[4]])
        outputPath = os.path.join(outputRoot, "回転" + str(degrees), "種類", "image", fruit, variety, name)
        outputTextPath = outputPath.replace("image", "annotation").replace(".jpg", ".txt")
    
    with open(outputTextPath, mode = 'w') as f:
            for oneCoorList in newCoorList:
                f.write(oneCoorList[0] + "," + oneCoorList[1] + "," + oneCoorList[2] + "," + oneCoorList[3] + "," + oneCoorList[4])
    rotatedImage.save(outputPath)
    

def makeFolder(root, folderName, imageFile):
    outputPath = os.path.join(root, folderName)
    if not os.path.exists(outputPath):
        os.mkdir(outputPath)
    
    outputPath = os.path.join(outputPath, "種類")
    if not os.path.exists(outputPath):
        os.mkdir(outputPath)
    
    outputAnnotationPath = os.path.join(outputPath, "annotation")
    if not os.path.exists(outputAnnotationPath):
        os.mkdir(outputAnnotationPath)
    outputImagePath = os.path.join(outputPath, "image")
    if not os.path.exists(outputImagePath):
        os.mkdir(outputImagePath)
    
    imagePathList = imageFile.split('\\')
    
    outputAnnotationPath = os.path.join(outputAnnotationPath, imagePathList[-3])
    if not os.path.exists(outputAnnotationPath):
        os.mkdir(outputAnnotationPath)
    outputImagePath = os.path.join(outputImagePath, imagePathList[-3])
    if not os.path.exists(outputImagePath):
        os.mkdir(outputImagePath)
    
    outputAnnotationPath = os.path.join(outputAnnotationPath, imagePathList[-2])
    if not os.path.exists(outputAnnotationPath):
        os.mkdir(outputAnnotationPath)
        outputAnnotationPath = os.path.join(outputAnnotationPath, 'classes.txt')
        with open(outputAnnotationPath, mode = 'w') as f:
            f.write("")
    outputImagePath = os.path.join(outputImagePath, imagePathList[-2])
    if not os.path.exists(outputImagePath):
        os.mkdir(outputImagePath)
    
            
```
